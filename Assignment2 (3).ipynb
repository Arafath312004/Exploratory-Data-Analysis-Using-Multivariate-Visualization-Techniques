```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv(r'C:\Users\Arafath\OneDrive\Desktop\processed_data news.csv')

# Set up the figure size and style for clarity
plt.figure(figsize=(16, 6))
sns.set(style="whitegrid")

# Plot 1: Price vs Storage with color-coded PPI Density
plt.subplot(1, 2, 1)
scatter1 = sns.scatterplot(
    data=data, x='storage', y='price_USD', hue='PPI_Density', palette='viridis', alpha=0.7, s=50
)
scatter1.set_title("Figure 1: Price (USD) vs Storage (GB) colored by PPI Density")
scatter1.set_xlabel("Storage (GB)")
scatter1.set_ylabel("Price (USD)")
plt.legend(title="PPI Density", loc="upper left")


```

```python
# Plot 2: Price vs PPI Density with color-coded Storage
plt.subplot(1, 2, 2)
scatter2 = sns.scatterplot(
    data=data, x='PPI_Density', y='price_USD', hue='storage', palette='viridis', alpha=0.7, s=50
)
scatter2.set_title("Figure 2: Price (USD) vs PPI Density colored by Storage")
scatter2.set_xlabel("PPI Density")
scatter2.set_ylabel("Price (USD)")
plt.legend(title="Storage (GB)", loc="upper left")

plt.tight_layout()
plt.show()
```

```python
# Trellis plot (FacetGrid) to observe Price vs Storage across different Years, colored by PPI Density
g = sns.FacetGrid(data, col="Year", hue="PPI_Density", palette="viridis", col_wrap=4, height=4, aspect=1)
g.map(sns.scatterplot, "storage", "price_USD", alpha=0.7, s=50)
g.add_legend(title="PPI Density")
g.set_axis_labels("Storage (GB)", "Price (USD)")
g.set_titles("Year: {col_name}")

plt.subplots_adjust(top=0.9)
g.fig.suptitle("Figure 3: Price (USD) vs Storage (GB) Across Different Years with PPI Density Color Coding")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns for the correlation matrix
numeric_data = data.select_dtypes(include='number')
correlation_matrix = numeric_data.corr()

# Set up the figure size and plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
